```python
# base de dados: https://docs.google.com/spreadsheets/d/1PNhGNtVK_FZWA59Dn7Kxnz6Dn9L-MalHPF6S3e0FbSQ/edit#gid=0
sheet_google=pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQQx8tapMpcdj0BUaRb8rWJQ1CbgEUU-YmLYd71pDsHPaWPkPTLfF5coZAr-cCz6OtsLa73ZF9UavYk/pub?output=csv')
dem_mkt = sheet_google.set_index('ticker')

def tab(dem_ju7):
  dem_ju7['CONTA_CDDS'] = dem_ju7['DS_CONTA'] + ' - ' + dem_ju7['CD_CONTA']
  dem_ju_atual = dem_ju7[dem_ju7['ANO'] == max(dem_ju7['ANO'])]
  dem_agrup7 = pd.pivot_table(dem_ju_atual[['ticker','segmento','CONTA_CDDS','VL_CONTA']],
                                  columns=['ticker','segmento'],
                                  values=['VL_CONTA'],
                                  index=['CONTA_CDDS'],
                                  fill_value='NaN')
  dem_tr7 = dem_agrup7.T
  dem_tr7.reset_index(level=0, inplace=True)
  del dem_tr7['level_0']
  dem_tr7 = dem_tr7.join(dem_mkt[['marketcap','volume']], how='outer').dropna() #Não uso número de ações (shares)
  dem_t7 = dem_tr7[dem_tr7['volume'] > 1000] #SE O GOOGLE FINANCE DER PROBLEMA, USA O BACKUP, COLUNA 'volume2', ATÉ NORMALIZAR O SISTEMA.
  return dem_t7.reset_index()

dem_ti7 = tab(dem_ind4)
dem_tc7 = tab(dem_con4)
# dem_ti7 = tab(dem_indt4)
# dem_tc7 = tab(dem_cont4)
```

```python
lista_patliq = ['patrimonio liquido conolidado - ' + c for c in ['2.08', '2.07', '2.03']]
lista_patliq.extend(['patrimonio liquido - ' + c for c in ['2.07', '2.05', '2.03']])
#-----------------------------------------------------------------------------------------------------------------------------------------------------------

lista_lucro = ['lucro prejuizo liquido conolidado periodo - ' + c for c in ['3.11']]
lista_lucro.extend(['lucro prejuizo liquido periodo - ' + c for c in ['3.11']])
lista_lucro.extend(['lucro prejuizo conolidado periodo - ' + c for c in ['3.13','3.11', '3.09']])
lista_lucro.extend(['lucro prejuizo periodo - ' + c for c in ['3.13','3.11']])
```

```python
# patliq = {'patliq':[]}
# lucro = {'lucro':[]}
# ticker = {'ticker':[]}
#-----------------------------------------------------------------------------------------------------------------------------------------------------------
#Pegando apenas os valores de dfp (não leva em conta os trimestrais)

patliq = []
lucro = []

for c in range(0,len(dem_tc7)):
  patrimonio = []
  find_pat = 0

  if dem_tc7['ticker'][c] == 'BRBI11':
    patliq.append(0)
    continue

  for pat in lista_patliq:
    if pat in dem_tc7.columns:
      if dem_tc7[pat][c] != 'NaN':
        patrimonio.append(dem_tc7[pat][c])
        find_pat +=1

  if find_pat == 0:
    for pat in lista_patliq:
      if pat in dem_ti7.columns:
        if dem_tc7[pat][c] != 'NaN':
          patrimonio.append(dem_ti7[dem_ti7['ticker']== dem_tc7['ticker'][c]][pat])

  if max(patrimonio) != 0:
    # patliq['patliq'].append(max(patrimonio))
    # ticker['ticker'].append(dem_tc7['ticker'][c])
    patliq.append(max(patrimonio))
  else:
    raise ValueError(f"Patrimônio líquido de {dem_tc7['ticker'][c]} é nulo!")

dem_tc7['patliq'] = patliq

#-----------------------------------------------------------------------------------------------------------------------------------------------------------
for c in range(0,len(dem_tc7)):

  lucr = []

  if dem_tc7['ticker'][c] == 'BRBI11':
    lucr.append(0)
    pass

  find_luc = 0
  for luc in lista_lucro:
    if luc in dem_tc7.columns:
      if dem_tc7[luc][c] != 'NaN':
        lucr.append(dem_tc7[luc][c])
        find_luc +=1

  if find_luc == 0:
    for luc in lista_lucro:
      if luc in dem_ti7.columns:
        if dem_tc7[luc][c] != 'NaN':
          lucro.append(dem_ti7[dem_ti7['ticker']== dem_tc7['ticker'][c]][luc])

  if max(lucr) != 0:
    # lucro['lucro'].append(max(lucr))
    # ticker['ticker'].append(dem_tc7['ticker'][c])
    lucro.append(max(lucr))
  else:
    raise ValueError(f"Lucro líquido de {dem_tc7['ticker'][c]} é nulo!")
dem_tc7['lucro'] = lucro

print('Todas as empresas foram consultadas! \n\n')
#-----------------------------------------------------------------------------------------------------------------------------------------------------------
print(f"Em caso de não localizar algum atributo, acesse o tipo o troca o CÓDIGO NUMÉRICO: https://sistemaswebb3-listados.b3.com.br/listedCompaniesPage/main/{int(codigosb3[codigosb3['ticker']=='BMGB4']['CD_CVM'])}/QUALQUERCOISA/reports?language=pt-br")

# dem_tc7['patliq'] = patliq['patliq']
# dem_tc7['lucro'] = lucro['lucro']

dem_tc7['ROE'] = dem_tc7['lucro'] / dem_tc7['patliq']
dem_tc7['L/P'] = dem_tc7['lucro'] / dem_tc7['marketcap']
```
